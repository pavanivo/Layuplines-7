Do a split/apply/combine: Group the dataset by the 'artist.inverted' column and apply the mean. What's the average rank that Usher sees in his first week?

Try grouping by 2 other columns. Write up any other analysis in a markdown cell

```python
billboard_df.groupby('artist.inverted').mean()
```

```python
billboard_df.groupby('genre').mean()
```

Usher got it bad compared to 98 degrees - he averages a rank of 68 in the first week. When grouping by genres, the Reggae genre sees the highest average 1st week ranking. Electronica sees the highest delta between weeks 1 and 2

Make a bar plot of the artists who had more than three tracks in the billboard charts in 2000. This will require that you groupby the artist (I suggest assigning this to a variable called "tracks_grouped_by_artist"):

```python
tracks_grouped_by_artist = billboard_df.groupby('artist.inverted')['track'].count()
```

Then run a vectorized operation, filtering for only those artists who had more than 3 tracks.

```python
tracks_grouped_by_artist = tracks_grouped_by_artist[tracks_grouped_by_artist > 3]
```

Then plot with a bar chart!

```python
tracks_grouped_by_artist.plot.bar(figsize=(20,20))
```
